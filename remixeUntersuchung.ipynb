```python
import pandas as pd
eins = pd.read_csv('clonesRemix1.csv', sep=';')
zwei = pd.read_csv('clonesRemix2.csv', sep=';')
drei = pd.read_csv('clonesRemix3.csv', sep=';')
vier = pd.read_csv('clonesRemix4.csv', sep=';')
fünf = pd.read_csv('clonesRemix5.csv', sep=';')
sechs = pd.read_csv('clonesRemix6.csv', sep=';')
sieben = pd.read_csv('clonesRemix7.csv', sep=';')
acht = pd.read_csv('clonesRemix8.csv', sep=';')
neun = pd.read_csv('clonesRemix9.csv', sep=';')
zehn = pd.read_csv('clonesRemix10.csv', sep=';')
```

```python
neu = pd.concat([eins,zwei], axis=0, ignore_index=True)
neu1 = pd.concat([neu,drei], axis=0, ignore_index=True)
neu2 = pd.concat([neu1,vier], axis=0, ignore_index=True)
neu3 = pd.concat([neu2,fünf], axis=0, ignore_index=True)
neu4 = pd.concat([neu3,sechs], axis=0, ignore_index=True)
neu5 = pd.concat([neu4,sieben], axis=0, ignore_index=True)
neu6 = pd.concat([neu5,acht], axis=0, ignore_index=True)
neu7 = pd.concat([neu6,neun], axis=0, ignore_index=True)
alle = pd.concat([neu7,zehn], axis=0, ignore_index=True)
```

```python
diagramm = alle.plot()
diagramm.set_xlabel("Projekte")
diagramm.set_ylabel("Anzahl der Klone")
diagramm
```

```python
fig = diagramm.get_figure()
fig.savefig("remix.pdf")
```
